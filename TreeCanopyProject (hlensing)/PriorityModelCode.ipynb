```python
import pandas as pd
import numpy as np
```

```python
file = pd.read_csv('PriorityCBG.csv', header=[0])
```

```python
df = pd.DataFrame(file)
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.replace({'UTC_quart': 'Q1', 'income_quart': 'Q1'}, .4)
df = df.replace({'UTC_quart': 'Q2', 'income_quart': 'Q2'}, .3)
df = df.replace({'UTC_quart': 'Q3', 'income_quart': 'Q3'}, .2)
df = df.replace({'UTC_quart': 'Q4', 'income_quart': 'Q4'}, .1)
```

```python
df['Total.Possible.Planting.Area.in.percent2'] = df['Total.Possible.Planting.Area.in.percent']/100
df['Possible.Vegetation.Planting.Area.in.percent2'] = df['Possible.Vegetation.Planting.Area.in.percent']/100
```

```python
df.head()
```

```python
df['priority_CBGs'] = (df['UTC_quart'] + df['Total.Possible.Planting.Area.in.percent2'] + df['Possible.Vegetation.Planting.Area.in.percent2'] + df['income_quart'])/2.8
```

```python
df = df.sort_values(by='priority_CBGs', ascending=False)
```

```python
df.head()
```

```python
df.columns
```

```python
df['Geographic.Area.Name'].head()
```

```python
df = df.drop(columns=['Unnamed: 0', 'X'])
```

```python
df.head()
```

```python
df.to_csv('Priority_CBGs_File.csv')
```
