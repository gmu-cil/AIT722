```python
import psycopg2
import json
import pandas as pd
from sqlalchemy import create_engine

# Establish a connection to the database by creating a cursor object
# The PostgreSQL server must be accessed through the PostgreSQL APP or Terminal Shell
# conn = psycopg2.connect(host="localhost", port = 5432, database="humanface", user="humanface", password="humanface")

df = pd.read_csv('./data/car2go_samples.csv')

```

```python
df.head()
```

```python
engine = create_engine('postgresql://humanface:humanface@localhost:5432/humanface')


```

```python
#migrating data from Pandas DF to PostgreSQL
df.to_sql('car2go', engine, if_exists='append', index=False)
```

```python
df = pd.read_sql_query("""SELECT * FROM car2go c WHERE c.fuel<60;""",con=engine)
df
```

```python
df = pd.read_sql_query("""SELECT * FROM car2go c WHERE c."smartPhoneRequired" IS NOT TRUE;""",con=engine)
df
```
